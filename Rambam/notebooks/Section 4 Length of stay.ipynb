# Figure 9

> A box plot of the LoS distribution by hour of the day. The blue diamonds are the means, while the black bars are the medians. (Use dataset 3.)

```python
plt.figure(figsize=(12, 6))
sns.boxplot(
    x="entry_hour", y="los_hrs", data=df[['entry_hour', 'los_hrs']], 
    showmeans=True, meanprops={"marker":"D", "markerfacecolor":"blue"},
    showfliers=False, whis=[25, 75], palette="pastel"
);

plt.xlabel('time of a day(hour)')
plt.ylabel('LOS (hour)');
```
